Una tienda registró ventas diarias durante una semana:



*  Lunes: 1200 soles

* Martes: 1500 soles
* Miércoles: 980 soles
* Jueves: 2100 soles
* Viernes: 1850 soles
* Sábado: 2300 soles
* Domingo: 1400 soles

Calcula el promedio diario, valores máximo/mínimo, desviación estándar y
total semanal.

```python
import numpy as np
dia = np.array(['Lunes','Martes','Miercoles','Jueves','Viernes','Sabado','Domingo'])
venta = np.array([1200,1500,980,2100,1850,2300,1400])
```

```python
print(f"Minimo {venta.min()} que corresponde al dia {dia[venta.argmin()]} ", )
print(f"Maximo {venta.max()} que corresponde al dia {dia[venta.argmax()]}")
print("Promedio", venta.mean())
print("Desviación estándar", venta.std())
print("Total semanal", venta.sum())
```

```python
union = np.stack((dia,venta),axis=1)
union
```

```python
ventas_unidas = union[:, 1].astype(int)
print(ventas_unidas)
print(f"Mínimo: {ventas_unidas.min()} que corresponde al día {union[ventas_unidas.argmin(), 0]}")
print(f"Máximo: {ventas_unidas.max()} que corresponde al día {union[ventas_unidas.argmax(), 0]}")
print("Promedio:", ventas_unidas.mean())
print("Desviación estándar:", ventas_unidas.std())
print("Total semanal:", ventas_unidas.sum())
```

Calcula ingresos, costos, utilidad y margen de utilidad.

```python
lista = np.array([[320,50,35],
                 [180,85,60],
                 [450,30,18],
                 [95,150,95],
                 [210,65,42]])
lista
```

```python
ingresos = (lista[:,0]*lista[:,1]).sum()
print(f'Ingresos Total: S/{ingresos}')
```

```python
#Ingresos
ingresos_p=np.array([]).astype(int)
for i in range(len(lista)):
  ingreso = (lista[i,0] * lista[i,1]).sum()
  print(f'El ingreso del Producto {i+1} es: {ingreso}')
  ingresos_p = np.append(ingresos_p,ingreso)
ingresos_p
```

```python
costos = (lista[:,0]*lista[:,2]).sum()
print(f'Costo Total: S/{costos}')
```

```python
#Costos
costos_p=np.array([]).astype(int)
for i in range(len(lista)):
  costo = (lista[i,0] * lista[i,2]).sum()
  print(f'El ingreso del Producto {i+1} es: {costo}')
  costos_p = np.append(costos_p,costo)
costos_p
```

```python
#Utilidad
utilidades=ingresos - costos
print(f'Utilidad Total: S/{utilidades}')
```

```python
utilidades_p=np.array([]).astype(int)
for i in range(len(lista)):
  utilidad = (ingresos_p[i]-costos_p[i])
  print(f'La utilidad del Producto {i+1} es {utilidad}')
  utilidades_p=np.append(utilidades_p,utilidad)
utilidades_p
```

```python
#Margen
margen = np.round(np.divide(utilidades,ingresos)*100,2)
print(f'Margen Total: {margen}%')
```

```python
margenes_p=np.array([]).astype(float)
for i in range(len(lista)):
  margen=np.round(np.divide(utilidades_p[i],ingresos_p[i])*100,2)
  print(f'La utilidad del Producto {i+1} es: {margen}%')
  margenes_p = np.append(margenes_p,margen)
margenes_p
```

En un call center, 6 empleados registraron llamadas durante 5 días. Calcula
promedio por empleado, mejor/peor desempeño, promedio general y totales
diarios.

```python
import numpy as np
call_center = np.array([[45,52,48,50,55],[38,42,40,39,44],[60,58,62,59,61],[42,45,43,46,44],[50,48,51,49,52],[35,38,36,37,39]])
call_center
```

```python
#PROMEDIO GENERAL
promedio_total = call_center.mean()
print(promedio_total)
```

```python
#PROMEDIO POR EMPLEADO
promedios=np.array([]).astype(float)
for i in range(len(call_center)):
  promedio = call_center[i].mean()
  print(f"El promedio de llamadas del Empleado {i+1} es {promedio}")
  promedios = np.append(promedios,promedio)
promedios
```

```python
#MEJOR EMPLEADO Y PEOR EMPLEADO
print(f"El mejor desempeño es el Empleado {promedios.argmax()+1} con desempeño {promedios.max()}")
print(f"El peor desempeño es el Empleado {promedios.argmin()+1} con desempeño {promedios.min()}")
```

```python
#TOTAL DE LLAMADAS
llamadas_totales = call_center.sum()
print(llamadas_totales)
```

```python
#TOTALES DIARIAS POR EMPLEADO
llamadas_diarias = np.array([]).astype(int)
for i in range(len(call_center)):
  llamada = call_center[i].sum()
  print(f"Total de llamadas del Empleado {i+1} es {llamada}")
  llamadas_diarias = np.append(llamadas_diarias,llamada)
llamadas_diarias
```
